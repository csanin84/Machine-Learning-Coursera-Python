## Importaciones

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

#plt.style.use('seaborn-whitegrid')
plt.style.use('ggplot')
```

## Funciones

```python
# graficar datos
def plotData(X,y):
    plt.plot(X, y, "rx", markersize = 10)
    plt.xlabel("Population of City in 10,000s")
    plt.ylabel("Profit in $10,000s")
    plt.show()
    
# calcular función de costos
def computeCost(X, y, theta):
    m = len(y)   
    h = np.dot(X,theta)
    e = (h - y)**2
    sum_e = np.sum(e)
    J = (0.5/m) * sum_e
    return J

# gradiente desendiente
def gradientDescent(X, y, theta, alpha, num_iter):
    # inicializar valores
    m = len(y)
    J_historico = np.zeros((num_iter))
    theta0_hist, theta1_hist = np.zeros((num_iter)),np.zeros((num_iter))
    for i in range(0, num_iter):
        factor = alpha/m
        h = np.dot(X,theta)
        derivada = np.dot(X.T,(h-y))        
        theta = theta - (factor * derivada)
        # Guardar el historico de costo   
        J_historico[i] = computeCost(X, y, theta);
        if theta.shape[0] == 2:
            theta0_hist[i]= theta[0][0]
            theta1_hist[i] = theta[1][0]
    
    return[ theta, J_historico, theta0_hist, theta1_hist ]

# Normalizar Matris de datos
#def featureNormalize(X):
```

## Parte 1 --- Graficando datos

```python
print("\nGraficando los datos...")

# cargar datos
datos = np.loadtxt(fname = "ex1data1.txt", delimiter =",")
X = datos[:,0]
y = datos[:,1]
m = len(y)
# convertir en vectores filas
X = X.reshape(m,1)
y = y.reshape(m,1)
plt.figure(figsize=(10,10))
plotData(X,y)
```

## Parte 2 --- Costo y Gradiente desendente 

```python
# adicionar columna de 1 a X
if X.shape[1] == 1:
    X = np.append(np.ones((m,1)), X, axis= 1)
# inicializar los parametros de ajuste theta
theta = np.zeros((2,1))

# estableciendo parametros del gradiente desendente
iteraciones = 1500
alpha = 0.01

# probando la función de costo con theta
J = computeCost(X, y, theta)
print("\nProbando la función de Costo...\n")
print("con theta = ", theta.T, "\nCosto calculado = {0:5.2f}".format(J))
print("Valor esperado del costo (approx) =  32.07")

# más tpruebas de la función de costo
J = computeCost(X, y, [[-1.], [2.]])
print("\nCon theta = [-1  2]\nCosto Calculado = {0:5.2f}".format(J))
print('Valor esperado del costo (approx) = 54.24\n')

# correr el gradiente desendiente
print("\nCorriendo gradiente desendiente ...\n")
theta, J_historico, theta0_hist, theta1_hist = gradientDescent(X, y, theta, alpha, iteraciones)

# imprimir los thetas
print("Theta encontrado por el gradiente desendiente = \n{} ".format(theta))
print("Valor esperado del costo (approx) = \n-3.6303\n1.1664\n")
print("\n")

# graficar el ajuste linear
plt.figure(figsize=(10,7))
plt.plot(X[:,1], y, "rx", markersize = 10, label = "Datos de Entrenamiento")
h = np.dot(X,theta)
plt.plot(X[:,1], h, "b-", markersize = 7, label = "Regresión Lineal" )
plt.xlabel("Población de la ciudad en 10,000s")
plt.ylabel("Ganancia en $10,000s")
plt.legend()
plt.show()

# Predecir valores para valores de población de 35000 y 70000
prediccion_1 = np.dot(theta.T,[[1],[3.5]])*10000
prediccion_2 = np.dot(theta.T,[[1],[7.]])*10000
print("Para una población = 35,000, nosotros predecimos una ganancia de: {0:5.2f}".format(prediccion_1[0][0]))
print("Para una población = 70,000, nosotros predecimos una ganancia de: {0:5.2f}".format(prediccion_2[0][0]))
```
